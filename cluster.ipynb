```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score
import json
```

```python
tfidf_matrix = json.load(open('tfidf_matrix','r'))
```

```python
sse = []
k_values = range(1,35)
for k in k_values:
    km = KMeans(n_clusters=k)
    km = km.fit(tfidf_matrix)
    sse.append(km.inertia_)
    
```

```python
plt.plot(k_values, sse, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
km = KMeans(n_clusters=45)
km = km.fit(tfidf_matrix)
km.inertia_
```

```python
print 'Sihouette Score - The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.'
```

```python
k_values = range(2,15)
silhouette_avg = []
for k in k_values:
    km = KMeans(n_clusters=k)
    km = km.fit_predict(tfidf_matrix)
    
    silhouette_avg.append(silhouette_score(tfidf_matrix, km))
```

```python
plt.plot(k_values, silhouette_avg, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Average')
plt.title('Silhouette Analysis For Optimal k')
plt.show()
```
